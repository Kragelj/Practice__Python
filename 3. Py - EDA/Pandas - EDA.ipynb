### EXPLORATORY DATA ANALYSIS in Pandas

Exploratory Data Analysis (EDA) is a critical step in understanding the structure, patterns, and relationships within a dataset before applying more complex modeling techniques.
<br>The Pandas library in Python provides powerful tools for EDA, allowing users to efficiently summarize, visualize, and clean data. Mastering EDA with Pandas enables users to transform raw data into valuable knowledge, setting the foundation for deeper analysis and predictive modeling.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\world_population_EDA.csv")
df
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' %x)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
df.sort_values(by='World Population Percentage', ascending=False).head(10)
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot = True)
plt.rcParams['figure.figsize'] = (20,7)
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot = True)
plt.rcParams['figure.figsize'] = (10,3)
plt.show()
```

```python
df.groupby('Continent').mean(numeric_only=True)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df[df['Continent'].str.contains('Oceania')].sort_values(by='2022 Population', ascending=False)
```

```python
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2
```

```python
df2.plot()
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```

```python
df.columns
```

```python
df2 = df.groupby('Continent')[['2022 Population', '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population', '1970 Population']].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
plt.rcParams['figure.figsize'] = (15,5)

df3.plot()
```

```python
df2 = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population',  '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2
```
